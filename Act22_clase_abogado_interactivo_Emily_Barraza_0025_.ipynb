#### Clase Abogado


```python
# Definimos la clase (el molde)
class Abogado:
    # El constructor: inicializa los atributos del jurista
    def __init__(self, nombre, especialidad):
        self.nombre = nombre
        self.especialidad = especialidad

    # Función (método) para que el abogado proteste
    def protestar(self):
        print(f"El Lic. {self.nombre} dice: ¡Objeción, su Señoría! ⚖️")

    # Función (método) para que el abogado cobre (su actividad favorita)
    def cobrar(self):
        print(f"El Lic. {self.nombre} está enviando la factura por honorarios. 💸")

    # Función (método) para que el abogado demande
    def demandar(self):
        print(f"El Lic. {self.nombre} está redactando una demanda contra la vecina. 📑")

# --- Creación de los objetos (instancias) ---

# Creamos al abogado penalista
penalista = Abogado("Harvey Specter", "Derecho Penal")

# Creamos a la abogada civil
civilista = Abogado("Alicia Florrick", "Derecho Civil")

# Creamos al abogado de dudosa procedencia
abogado_turbio = Abogado("Saul Goodman", "Chicanas Legales")

# --- Uso de los objetos ---

print(f"Abogado 1: {penalista.nombre}, Especialidad: {penalista.especialidad}")
penalista.protestar()
penalista.cobrar()
penalista.demandar()
print("-" * 30)

print(f"Abogado 2: {civilista.nombre}, Especialidad: {civilista.especialidad}")
civilista.protestar()
civilista.cobrar()
civilista.demandar()
print("-" * 30)

print(f"Abogado 3: {abogado_turbio.nombre}, Especialidad: {abogado_turbio.especialidad}")
abogado_turbio.protestar()
abogado_turbio.cobrar()
abogado_turbio.demandar()
print("Clase abogado elaborada por Emily Barraza NL = 7")
```
